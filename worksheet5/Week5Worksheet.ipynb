```python
np.random.seed(42)

indices = np.arange(len(X))
np.random.shuffle(indices)

split = int(0.8 * len(X))

train_idx = indices[:split]
test_idx = indices[split:]

X_train = X[train_idx]
X_test  = X[test_idx]

Y_train = Y[train_idx]
Y_test  = Y[test_idx]

print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
def cost_function(X, Y, W):
    n = len(Y)
    Y_pred = np.dot(X, W)
    cost = (1 / (2 * n)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

```python
X_test = np.array([[1, 2], [3, 4], [5, 6]])
Y_test = np.array([3, 7, 11])
W_test = np.array([1, 1])

cost = cost_function(X_test, Y_test, W_test)

if cost == 0:
    print("Proceed Further")
else:
    print("Something went wrong")
```

```python
def gradient_descent(X, Y, W, alpha, iterations):

    m = len(Y)
    cost_history = []

    for _ in range(iterations):
        Y_pred = np.dot(X, W)
        loss = Y_pred - Y
        dw = (1 / m) * np.dot(X.T, loss)
        W = W - alpha * dw
        cost = cost_function(X, Y, W)
        cost_history.append(cost)

    return W, cost_history
```

```python
np.random.seed(0)

X = np.random.rand(100, 3)
Y = np.random.rand(100)
W = np.random.rand(3)

alpha = 0.01
iterations = 1000

final_params, cost_history = gradient_descent(X, Y, W, alpha, iterations)

print("Final Parameters:", final_params)
print("Final Cost:", cost_history[-1])
```

```python
def rmse(Y, Y_pred):
    return np.sqrt(np.mean((Y - Y_pred) ** 2))
```

```python
def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)
```

```python
def main():

    data = pd.read_csv('/content/student.csv')

    X = data[['Math', 'Reading']].values
    Y = data['Writing'].values

    np.random.seed(42)
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    split = int(0.8 * len(X))

    X_train = X[indices[:split]]
    X_test  = X[indices[split:]]
    Y_train = Y[indices[:split]]
    Y_test  = Y[indices[split:]]

    W = np.zeros(X_train.shape[1])
    alpha = 0.00001
    iterations = 1000

    W_opt, cost_history = gradient_descent(X_train, Y_train, W, alpha, iterations)

    Y_pred = np.dot(X_test, W_opt)

    print("Final Weights:", W_opt)
    print("RMSE:", rmse(Y_test, Y_pred))
    print("R2 Score:", r2(Y_test, Y_pred))


if __name__ == "__main__":
    main()
```
