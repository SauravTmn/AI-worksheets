```python
#TASK1
import numpy as np

print("\n=== PROBLEM 1 ===")

# 1
a = np.empty((2,2))
print("\n1. Empty array (2x2):")
print(a)

# 2
b = np.ones((4,2))
print("\n2. Ones array (4x2):")
print(b)

# 3
c = np.full((3,3), fill_value=7)
print("\n3. Full array of 7s:")
print(c)

# 4
given = np.array([[1,2,3],[4,5,6]])
d = np.zeros_like(given)
print("\n4. zeros_like(given):")
print(d)

# 5
e = np.ones_like(given)
print("\n5. ones_like(given):")
print(e)

# 6
new_list = [1,2,3,4]
f = np.array(new_list)
print("\n6. List → numpy array:")
print(f)

```

```python
#TASK2
print("\n=== PROBLEM 2 ===")

# 1
p1 = np.arange(10, 50)
print("\n1. Array 10–49:")
print(p1)

# 2
p2 = np.arange(9).reshape(3,3)
print("\n2. 3x3 matrix (0–8):")
print(p2)

# 3
p3 = np.eye(3)
print("\n3. Identity matrix:")
print(p3)

# 4
p4 = np.random.random(30)
print("\n4. Random array (30 values):")
print(p4)
print("Mean =", p4.mean())

# 5
p5 = np.random.random((10,10))
print("\n5. 10x10 random array:")
print(p5)
print("Min =", p5.min(), "Max =", p5.max())

# 6
p6 = np.zeros(10)
p6[4] = 1
print("\n6. Zero array with 5th element = 1:")
print(p6)

# 7
arr = np.array([1,2,0,0,4,0])
p7 = arr[::-1]
print("\n7. Reversed array:")
print(p7)

# 8
p8 = np.ones((6,6))
p8[1:-1, 1:-1] = 0
print("\n8. 2D array with border 1 and inside 0:")
print(p8)

# 9
p9 = np.zeros((8,8))
p9[1::2, ::2] = 1
p9[::2, 1::2] = 1
print("\n9. Checkerboard pattern:")
print(p9)
```

```python
#TASK3
print("\n=== PROBLEM 3 ===")

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

# 1
print("\n1. x + y:")
print(x + y)

# 2
print("\n2. x - y:")
print(x - y)

# 3
print("\n3. x * 3:")
print(x * 3)

# 4
print("\n4. Square of x:")
print(x**2)

# 5
print("\n5. Dot products:")
print("v·w =", np.dot(v,w))
print("x·v =", np.dot(x,v))
print("x·y =\n", np.dot(x,y))

# 6
print("\n6. Concatenate x & y (row-wise):")
print(np.concatenate((x,y), axis=0))

print("\nConcatenate v & w (column-wise):")
print(np.vstack((v,w)))

# 7
print("\n7. Why x & v fails:")
print("Error: x is 2D but v is 1D → shapes don’t match.")

```

```python
#TASK4
print("\n=== PROBLEM 4 ===")

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1
A_inv = np.linalg.inv(A)
print("\n1. A × A⁻¹ = Identity:")
print(A.dot(A_inv))

# 2
AB = A.dot(B)
BA = B.dot(A)
print("\n2. AB:")
print(AB)
print("\nBA:")
print(BA)
print("\nAB != BA confirmed")

# 3
print("\n3. (AB)^T == B^T × A^T:")
print("LHS:\n", AB.T)
print("RHS:\n", B.T.dot(A.T))

# Solving system of equations
A_sys = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])
B_sys = np.array([-1, -3, 9])

solution = np.linalg.inv(A_sys).dot(B_sys)
print("\nSolution to system [x, y, z]:")
print(solution)
```
